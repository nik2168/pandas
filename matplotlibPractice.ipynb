```python
import matplotlib.pyplot as plt
import numpy as np  
import pandas as pd
```

### Line Gaphs

```python
# line graph
x = np.array([2022, 2023, 2024, 2025]) # years
y = np.array([45, 75, 34, 55]) # count of student each year

# plot
plt.plot(x, y) # x and y coordinates

# labels
plt.title('Line Graph')
plt.xlabel('Years')
plt.ylabel('Number of Students')

# final draw of graph
plt.show() # to draw the graph
```

### Football Data Graphs

```python
df = pd.read_csv('./SP1.csv')
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
df.head()
```

Histogram

```python
hs_team = df['HS'].dropna() # drop missing values
hs_team

plt.figure(figsize=(10, 6))
plt.hist(hs_team, bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Home Team Goals')
plt.xlabel('Goals')
plt.ylabel('Count')
plt.show()
```
